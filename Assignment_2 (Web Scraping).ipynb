```python
# Importing Libraries

import requests
from bs4 import BeautifulSoup
import pandas as pd
import bs4
```

```python
#Use requests library and BeautifulSoup to connect to http://quotes.toscrape.com/ and get the HMTL text from the homepage.
```

```python
# URL that needs to be Scrapped

url = "https://quotes.toscrape.com/"
```

```python
response = requests.get(url)
```

```python
response
```

```python
response = response.content
```

```python
response
```

```python
soup = BeautifulSoup(response,'html.parser')
```

```python
soup
```

```python
#Get the names of all the authors on the first page.
```

```python
authors = set()
author_tags = soup.select('.author')
for author in author_tags:
    authors.add(author.text)
```

```python
authors
```

```python
#Create a list of all the quotes on the first page.
```

```python
quotes = []
text_tags = soup.select('.text')
for quote in text_tags:
    quotes.append(quote.text)
```

```python
quotes
```

```python
#Inspect the site and use Beautiful Soup to extract the top ten tags from the requests text shown on the top right from the home page (e.g Love,Inspirational,Life, etc...)
```

```python
top_ten_tags = []
tags = soup.select('.tag-item')

for tag in tags:
    top_ten_tags.append(tag.text)
```

```python
for tag in top_ten_tags:
    print(tag)
```

```python
#Used loop to scrape for all the 10 Pages
```

```python
author = []
base_url = 'http://quotes.toscrape.com/page/{}/'
current_page = 1
authors = set()

while True:
    res = requests.get(base_url.format(current_page))
    soup = bs4.BeautifulSoup(res.text, 'lxml')
    author_tags = soup.select('.author')
    
    if len(author_tags) == 0:
        break
    
    for author in author_tags:
        authors.add(author.text)

    current_page += 1

for author in authors:
    print(author)
```
